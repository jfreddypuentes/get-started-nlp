```python
import pandas as pd
```

### Read data

```python
data = pd.read_csv('base.csv')
data.columns = ['row_num', 'student_name', 'answer', 'os']
data.head()
```

### Contar el número de palabras por estudiante

```python
words_quantity = [len(answer) for answer in data["answer"].str.split(" ")]
data['words_quantity'] = words_quantity
```

```python
data[['student_name', 'answer', 'words_quantity']].sort_values(by=['words_quantity'], ascending=False)
```

### Hallar la palabra más utilizada en todo el conjunto de datos

```python
pd.Series(' '.join(data['answer']).split()).value_counts()[:30]
```
